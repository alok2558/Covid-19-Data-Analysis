```python
# total population of a state were subtarcted from vaccinated population
df['populationleft'] = df['population']  - df['vaccination']
```

```python
df['date'] = 0
```

```python
import math
# datetime imported for converting a given number of days to a particular date
import datetime
# for each state we divided left population by rate of vaccination and the result is added to 14 aug 2021 ,to get the desired date
for i in df.State.unique():
    indexname = df[df['State'] == i].index
    x = df.loc[indexname,'populationleft'].values
    x=x[0]
    y = df.loc[indexname,'rate_vac'].values
    y = y[0]
    day = x/y
    day = math.ceil(day)
    start_date = datetime.date(2021,8,14)
    delta = datetime.timedelta(days=day)
    end_date = start_date + delta
    df.loc[indexname,'date'] = end_date
```
